```python
from sentence_transformers import SentenceTransformer
model = SentenceTransformer('sentence-transformers/bert-base-nli-mean-tokens')
```

```python
import numpy as np
```

```python
sentence1 = "i love singing western"
sentence2 = "i sing western"
embedd1 = model.encode(sentence1)
embedd2 = model.encode(sentence2)
# print(embeddings.shape)
```

```python
unit_vector1 = embedd1/np.linalg.norm(embedd1)
unit_vector2 = embedd2/np.linalg.norm(embedd2)
```

```python
sim = np.dot(unit_vector1,unit_vector2)
print(sim)
```

```python
sentence = """At eight o'clock on wednesday morning, Arthur didn't feel very good. So i took him to doctor."""

sentence = sentence.lower()

word_tokens = nltk.word_tokenize(sentence)
sent_tokens = nltk.sent_tokenize(sentence)
print(word_tokens)
print(sent_tokens)

res = ""
stemmer = PorterStemmer()
for token in word_tokens:
   res += " "+ stemmer.stem(token)
  
print(res)
```

```python
def preprocess(sentence):
    sentence = sentence.lower()
    word_tokens = nltk.word_tokenize(sentence)
    res = ""
    stemmer = PorterStemmer()
    for token in word_tokens:
        res += " "+ stemmer.stem(token)
    print(res)
    return res
```

```python
sentence1 = "COVID‑19 transmits when people breathe in air contaminated by droplets and small airborne particles containing the virus. The risk of breathing these in is highest when people are in close proximity, but they can be inhaled over longer distances, particularly indoors. Transmission can also occur if splashed or sprayed with contaminated fluids in the eyes, nose or mouth, and, rarely, via contaminated surfaces. People remain contagious for up to 20 days, and can spread the virus even if they do not develop symptoms"
```

```python
s1 = preprocess(sentence)
```

```python
s2 = preprocess("i took arthur to doctor at 8 o'clock in the morning on wednesday")
```

```python
e1 = model.encode(s1)
e2 = model.encode(s2)
sim = np.dot(e1,e2)/(np.linalg.norm(e1)*np.linalg.norm(e2))
print(sim)
```
